## Part V: Radiation heat transfer

Now implement radiation heat transfer at the battery surface (anode and cathode bondaries).
\begin{equation}
    \dot{Q}^{\prime\prime}_{\rm rad} = \sigma\epsilon\left(T^4_{\rm amb} - T^4_{\rm surf}\right)\frac{A}{V}
\end{equation}
The model code has already defined `sigma`, the Stefan-Boltzmann constant, plus `pars.emmissivity`, the surface emmissivity ($\epsilon$), the ambient temperature `pars.T_amb`, and `pars.A_ext`, the surface area per unit volume for both electrodes, (i.e. $\frac{A}{V}$ in the equation above. Tthe same value is used for both electrodes).

Calculate `Q_rad`, the total heat transferred to each component per unit volume.


```python
thermal_flags.rad = 1
cycle_function(thermal_flags)
```

### Discussion 

<font color='red'>[**Discuss the results here**]</font>

What do you notice?  Is this believable?  Why do some C-rates reach a steady-state value, and some do not?

For those that do reach a steady-state temperature, what determines the steady state value?  What processes are being balanced, at steady state?

At steady state the heat generation of the cell is being balanced by radiative heat flux away from the cell surface. At low C-rates the heat generation rate is so low that even a small temperature over that of the ambient creates a high enough radiant heat flux to compensate. At higher C-rates, however, the rate of heat generation becomes so much faster that a higher temperature difference between cell and ambient is required to balance the heat fluxes. At 10C, this does not occur below 100 °C.

For thicker batteries (e.g. stack of pouch cells or large-format spiral-wound cells), the external area per volume will be lower, the radiative flux from the surface needs to be balanced by a conductive flux within the cell - from the center towards the cell surface. This could lead to more significant temperature gradients and "hot spots" inside the battery.

3. Discuss the influence of C-rate on our battery's thermal response.  Is the relationship between C-rate and max temperature linear? (hint: it is not) Why do you think this is? <font color='red'>[**Discuss here**]</font>

The max-temp response isn't linear because the heat-generation as a function of current density isn't linear. For example, the electron enthalpy depends on overpotential, which is a logarithmic function of current density. Ohmic heating, on the other hand, depends on $I^2$. 

The temperature increase is faster in absolute terms (e.g. °C/s) for higher C-rates. The rate of temperature increase is proportional to the difference between the heat generation flux (reaction, ohmic heating, thermal conduction) and the heat dissipation flux (radiation and convection). The generation flux increases with C-rate but is not strongly dependent on time, while the dissipation flux is strongly time-dependent. I think this is why it takes a larger fraction of discharge capacity to reach a steady-state temperature at higher C-rates - it takes longer for the dissipation flux to equalize with the higher generation flux.
